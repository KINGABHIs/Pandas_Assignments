Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
scatter_plot = px.scatter(titanic, x = 'age',y='fare',title= 'scatter plot of age vs Fare',labels={'age':'Age','fare':'Fare'},template='plotly_white',color="class")
scatter_plot.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

A distplot is short for distribution,or visualizes the distribution of a dataset. It combines a histogram with a kernel density estimate to show the shape and spread of the data. Distplots are useful for understanding the underlying distribution, central tendency, and variability of a dataset.

Using Plotly Express for a Distribution Plot:
While Plotly Express doesn't have a direct distplot function (like Seaborn), you can create a histogram with KDE enabled to achieve the same effect.

Here is how you can create a distribution plot using Plotly Express:

```python
import plotly.express as px

tips = px.data.tips()
tips
```

```python

distplot = px.histogram(tips, x='total_bill',marginal='box', nbins=30, title='Distribution plot of tolal bill',
                        labels={'total_bill':"Total Bill"},
                        opacity=0.7,
                        histnorm ='density')
distplot.show()
```
