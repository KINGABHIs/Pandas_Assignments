# Seaborn Assignment

```python
import seaborn as sns

data = sns.load_dataset("fmri")
data
```

```python
sns.lineplot(x = data.timepoint, y = data.signal, data = data)
```

```python
sns.lineplot(data=data, x="timepoint", y="signal", hue="event", style="region", markers=True)
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.boxplot(data = titanic, x = "pclass", y = "age")
```

```python
sns.boxplot(data = titanic, x = "pclass", y = "fare")
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
sns.histplot(data = diamonds ,x = diamonds.price, hue = 'cut', multiple= "stack", kde= False ,bins=30)
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.pairplot(data=iris, hue ="species", height=2.5, diag_kind="kde")
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot = flights.pivot(index ="month", columns= "year", values ="passengers")
sns.heatmap(data=flights_pivot, annot=True, fmt='d', cmap='YlGnBu', linewidths=.5)
```
