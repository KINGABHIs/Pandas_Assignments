```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x, y, color='blue', alpha=0.7)
plt.title("Scatter Plot of Generated Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
x1, y1 = np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500])
x2, y2 = np.array([0, 1, 2, 3, 4, 5]), np.array([50, 20, 40, 20, 60, 70])
x3, y3 = np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60])
x4, y4 = np.array([0, 1, 2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(10, 8))  # Adjust figure size
plt.subplot(2, 2, 1)  # Grid of 2x2, position 1
plt.plot(x1, y1, color='blue', marker='o')
plt.title("Line 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2, 2, 2)  # Grid of 2x2, position 2
plt.plot(x2, y2, color='green', marker='s')
plt.title("Line 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2, 2, 3)  # Grid of 2x2, position 3
plt.plot(x3, y3, color='red', marker='^')
plt.title("Line 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.subplot(2, 2, 4)  # Grid of 2x2, position 4
plt.plot(x4, y4, color='purple', marker='d')
plt.title("Line 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.tight_layout()
plt.show()
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10, 5))
plt.bar(company, profit, color='skyblue')
plt.title("Profit by Company")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.barh(company, profit, color='lightgreen')
plt.title("Profit by Company (Horizontal)")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```
